```python

# Assuming 'ejection_fraction' is a column in your DataFrame 'data'
fig, ax1 = plt.subplots(1, 1, figsize=(24, 8))
sns.barplot(x=data['age'].value_counts().index, y=data.groupby('age')['ejection_fraction'].mean(), ax=ax1)
plt.title('Relationship between Age and Ejection Fraction')
plt.xlabel('Age')
plt.ylabel('Ejection Fraction')
plt.show()
```
